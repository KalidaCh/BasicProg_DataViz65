3. จำนวนที่จอดรถและห้องน้ำของของทรัพย์สินที่ถูกที่สุดแต่ละชนิด


```python
#เปรียบเทียบราคาเฉลี่ยของ Detached house, Condo, Twin house และ Townhome
unit.groupby(['propertytype_name_en'])['price_min'].describe()
```

```python
unit.groupby(['propertytype_name_en'])['price_min'].mean() #ค่าเฉลี่ยใส่ .mean()
```

```python
unit.groupby(['propertytype_name_en'])['price_min'].mean()[[ 'Detached House','Condo','Twin house','Townhome']]
```

```python
#พื้นที่ (area_total_min_wa และ area_usable_min) และ จำนวนห้องนอน ของทรัพย์สินที่แพงที่สุดแต่ละชนิด
unit.groupby(['propertytype_name_en'])[[ 'area_total_min_wa','area_usable_min','count_room_bed','price_min']].max()
```

```python
#3. จำนวนที่จอดรถและห้องน้ำของของทรัพย์สินที่ถูกที่สุดแต่ละชนิด
unit.groupby(['propertytype_name_en'])[[ 'count_parking','count_room_bath','price_min']].min()
```
