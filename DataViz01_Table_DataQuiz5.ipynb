# Parallel Coordinates  

```python
pd.plotting.parallel_coordinates(df, 'Name') # inputตัวแรก df คือชื่อตารางที่โหลดมา inputตัวที่ 2 Name คือจัดกรุ้ปตามคอลัมน์ Name  plotคอลัมน์ที่ไม่ได้เป็นกรุ้ป เอาเนมที่ใส่เข้าไปมาแยกสี
```

แปลผล ดอกเล็กๆสีเขียวลักณะคล้ายๆกันกลีบเลี้ยงกลีบดอกจะยาว ส่วนสีม่วงกลีบดอกเล็ก เราแยกดอกsentosa ออกจากดอกที่เหลือได้

### select columns to show

```python
pd.plotting.parallel_coordinates(df, 'Name',cols = ['SepalLength','PetalLength'])  #ดูแค่SepalLength,PetalLength ใส่ cols = ['SepalLength','PetalLength']) 
                                                                                   #ถ้าอยากเอาอันไหนมาเปรียบเทียบกันก็สลับได้
```

```python
pd.plotting.parallel_coordinates(df, 'Name',cols = ['SepalLength','PetalLength','SepalWidth','PetalWidth'])  
```

```python
pd.plotting.parallel_coordinates(df, 'Name',cols = ['SepalLength','PetalLength']);
```

### change color

https://matplotlib.org/stable/tutorials/colors/colors.html

```python
pd.plotting.parallel_coordinates(df, 'Name', color=['c','m','k']) #เปลี่ยนสีดูในลิงก์ข้างบน ใช้ตัวย่อของชื่อสี สีของเเสง #สีของเม็ดสีที่เขาเอามาผสมแบบปริ้นๆ เห็นตามหมึกพิมพ์ต่างๆ
```

```python
pd.plotting.parallel_coordinates(df, 'Name', color=['r','y','k']) 
```

```python
pd.plotting.parallel_coordinates(df, 'Name', color=[[0.5,0.5,0.5],[0,0,0.8],[0.2,0.4,0.7]]) # 1 คือปล่อยสีนั้นเต็มๆ 0คือไม่ปล่อยเลย 
```

```python
pd.plotting.parallel_coordinates(df, 'Name', color=[[0.4,0,0],[0,0,0.8],[0.2,0.4,0.7]]) #RGB แสงผสมกัน
```

```python
pd.plotting.parallel_coordinates(df, 'Name', color=['#FF00E0', '#00FFE2', '#88B46C']) #html color เปลี่ยนสีล่างสุดเป็นสีมข html  สีตรงกลางเป็นสีที่ชอบ
```

```python
matplotlib.rcParams['figure.figsize']=[10,10]
```

```python
plt.subplot(1,3,1)  # subplot(มีกี่แถว, มีกี่หลัก, เป็นตัวที่เท่าไหร่)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot')

plt.subplot(1,3,2)
plt.imshow(df.iloc[50:100,:-1],interpolation='nearest',cmap='hot')

plt.subplot(1,3,3)
plt.imshow(df.iloc[100:150,:-1],interpolation='nearest',cmap='hot')
```

```python
plt.subplot(1,4,1)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot')

plt.subplot(1,4,2)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot',alpha = 0.2)

plt.subplot(1,4,3)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot',alpha = 0.5)

plt.subplot(1,4,4)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot',alpha = 0.8)
```
