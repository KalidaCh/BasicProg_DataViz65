```python
unit = pd.read_csv('/content/drive/MyDrive/BPDV65/opendata_unittype.csv' )
unit.head()

```

```python
project = pd.read_csv('/content/drive/MyDrive/BPDV65/opendata_project.csv')
project.head()
```

```python
unit1=unit.dropna(subset=['propertytype_name_th','price_min','area_usable_min'])
unit1

```

```python
project1=project.dropna(subset=['province_name_th'])
project1

```

```python
unitandproject = unit1.merge(project1, left_on='row_number', right_on='row_number')
unitandproject .head()

```

```python
unitandproject1 = unitandproject [unitandproject ['province_name_th']=='เชียงใหม่'] #เอามาแค่เชียงใหม่
unitandproject1

```

```python
home =  unitandproject1[unitandproject1['propertytype_name_th_x']=='บ้าน']
home

```

```python
condo =  unitandproject1[unitandproject1['propertytype_name_th_x']=='คอนโด']
condo

```

```python
HomeandCondo = pd.concat([home,condo]) 
HomeandCondo 

```

```python
HomeandCondo['propertytype_name_th_x'].unique()

```

```python
pd.plotting.parallel_coordinates(HomeandCondo, 'propertytype_name_th_x',cols = ['price_min_x','area_usable_min'], color=['#3D1C02','#FF81C0'])
```
